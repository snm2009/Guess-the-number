```python
import random

number = random.choice(range(0, 101, 2))
print("I am thinking of an even number between 0 and 100.")

while True:
    guess = int(input("Enter your guess: "))

    if guess < number:
        print("Too low!")
    elif guess > number:
        print("Too high!")
    else:
        print("Congratulations! You guessed the number!")
        break
```
